```python
from pyspark.sql.functions import current_timestamp
```

```python
input_path= "/Volumes/my_tranings_demos/bronze/data_bucket/retail_customer_source_files/"

checkpoint = "/Volumes/my_tranings_demos/bronze/data_bucket/retail_customer_source_files/sales_checkpoint"

output_table = "my_tranings_demos.bronze.bronze_sales"

(
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("cloudFiles.includePattern","sales.*\\.csv")
    .option("cloudFiles.schemaLocation", checkpoint)
    .option("header","true")
    .option("cloudFiles.schemaEvolutionMode", "rescue")
    .load(input_path)
    .withColumn("load_time", current_timestamp())
    .writeStream
    .option("checkpointLocation", checkpoint)
    .trigger(availableNow=True)
    .toTable(output_table)
)
```
